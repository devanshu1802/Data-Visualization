```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import folium

data = pd.DataFrame({
    'city': ['New York', 'Los Angeles', ' Chicago', 'Houston', 'Phoenix'],
    'lat' : [40.7128, 34.0522, 41.8781, 29.7604, 33.4484],
    'lon' : [-74.0060 , -118.2437, -87.6298, -95.3698, -112.0740]
})
```

```python
m = folium.Map(location=(45.5236, -122.6750),
               zoom_start=8,
               zoom_control=True)

#Created a map centered at Portland, Oregon with a zoom level of 4
#folium.Map parameters are = location (latitude, longitude) you can find there coordinates by searching google map
#Adding points to the map

for index, row in data.iterrows():
    folium.Marker([row['lat'], row['lon']],
                  popup=row['city'],
                  tooltip="Click Me !",
                  icon=folium.Icon(color="black")).add_to(m)
    
m.save("map.html")
m    
```

```python
d = pd.DataFrame({
    'State Name' : ['Delhi', 'Mumbai', 'Punjab', 'Haryana', 'Goa', 'Uttarakhand', 'Uttar Pradesh'],
    'Latitude' : [28.7041, 19.0760, 31.1471, 29.0588, 15.2993, 30.0668, 26.8467],
    'Longitude' : [77.1025, 72.8777, 75.3412, 76.0856, 74.1240, 79.0193, 80.9462],
    "Sales" : [150,200,130,120,90,80,170]
})

chart = folium.Map(location=[15.043199, 36.9921626],
                   zoom_start=5,
                   zoom_control=True)

colorchart = folium.LinearColormap(['green','yellow','orange','red'],
                                   vmin=d["Sales"].min(),
                                   vmax=d["Sales"].max(),
                                   caption="Sales Amount")

for column, row in d.iterrows():
    folium.CircleMarker(
        location=[row["Latitude"], row["Longitude"]],
        radius=row["Sales"] / 20,
        color = colorchart(row["Sales"]),
        fill=True,
        fill_color= colorchart(row["Sales"]),
        fill_opacity=0.7,
        popup=f"{row['State Name']}: {row['Sales']}M").add_to(chart)

chart.save("map.html")
chart    
```

```python
indian_sales = pd.DataFrame({
    "State Name":['Delhi','Mumbai','Punjab','Haryana','Goa','Uttarakhand','Uttar Pradesh'],
    "Latitude":[28.7041,19.0760,31.1471,29.0588,15.2993,30.0668,26.8467],
    "Longitude":[77.1025,72.8777,75.3412,76.0856,74.1240,79.0193,80.9462],
    "Sales":[150,200,130,120,90,80,170],
    "Units Sold":[12000,18000,11000,9000,7000,6500,15000]
})
```

```python
chart = folium.Map(location=[15.3494864,73.6770313],
               zoom_start=5,
               zoom_control=True,
              tiles="CartoDB dark_matter")
colorchart = folium.LinearColormap(colors=["green","yellow",'orange','red'],
                              vmin=indian_sales["Sales"].min(),
                              vmax=indian_sales["Sales"].max(),
                              caption="Sales Amount")
for column, row in indian_sales.iterrows():
    folium.CircleMarker(
        location=[row["Latitude"], row["Longitude"]],
        radius=row["Sales"] / 20,
        color=colorchart(row["Sales"]),
        fill=True,
        fill_color=colorchart(row["Sales"]),
        fill_opacity=0.7,
        popup=f"{row['State Name']}: Rs{row['Sales']}M"
        ).add_to(chart)
    colorchart.add_to(chart)
chart
```
